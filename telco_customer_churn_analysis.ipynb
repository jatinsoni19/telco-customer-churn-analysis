Task 2: Data Cleaning


```python
# See missing values again
df.isnull().sum()
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
# Remove Duplicates
df = df.drop_duplicates()
```

Task 3: Exploratory Data Analysis

```python
# Summary Statistics
df.describe()
```

```python
# Churn Distribution
import matplotlib.pyplot as plt

df['churn'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Churn vs Non-Churn')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Histogram for Monthly Charges
df['monthlycharges'].plot(kind='hist', bins=20, color='blue')
plt.title('Monthly Charges Distribution')
plt.xlabel('Monthly Charges')
plt.show()
```

```python
# Boxplot for Tenure
import seaborn as sns

sns.boxplot(x=df['tenure'])
plt.title('Tenure Boxplot')
plt.show()
```

Task 4: Customer Segmentation Visualization

```python
# Create Tenure Categories
def tenure_group(tenure):
    if tenure <= 12:
        return '0-12 months'
    elif tenure <= 36:
        return '13-36 months'
    else:
        return '37+ months'

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
# Pie Chart of Tenure Groups
df['tenure_group'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ff9999'])
plt.title('Customer Distribution by Tenure')
plt.ylabel('')
plt.show()
```

```python
# Bar Chart of Avg Monthly Charges per Tenure Group
avg_charges = df.groupby('tenure_group')['monthlycharges'].mean()

avg_charges.plot(kind='bar', color='salmon')
plt.title('Avg Monthly Charges by Tenure Group')
plt.ylabel('Average Monthly Charges')
plt.xlabel('Tenure Group')
plt.xticks(rotation=0)
plt.show()
```

Task 5: Advanced Analysis

```python
# Group Customers by Tenure - Charges & Churn Stats
grouped = df.groupby('tenure_group').agg({
    'monthlycharges': 'mean',
    'churn': lambda x: (x == 'Yes').mean() * 100  # % of churned customers
})

print(grouped)
```

```python
# Churn by Demographics
# By Gender
gender_churn = df.groupby('gender')['churn'].value_counts(normalize=True).unstack() * 100
print(gender_churn)
```

```python
#  By Senior Citizen
senior_churn = df.groupby('seniorcitizen')['churn'].value_counts(normalize=True).unstack() * 100
print(senior_churn)
```

```python
# By Contract Type
contract_churn = df.groupby('contract')['churn'].value_counts(normalize=True).unstack() * 100
print(contract_churn)
```
